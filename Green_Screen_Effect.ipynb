```python
import cv2 
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
```

```python
image1=cv2.imread('/content/drive/MyDrive/Computer vision Images/astronaut.jpg')
```

```python
cv2_imshow(image1)
```

```python
print('Image Type:',type(image1))
print('Image dimensions:',image1.shape)
```

```python
image_cp=np.copy(image1)
```

```python
image_cp=cv2.cvtColor(image_cp,cv2.COLOR_BGR2RGB)
cv2_imshow(image_cp)
```

```python
lowerGreen=np.array([0,100,0])
upperGreen=np.array([120,255,100])
```

```python
mask=cv2.inRange(image_cp,lowerGreen,upperGreen)
cv2_imshow(mask)
```

```python
masked_image=np.copy(image_cp)
masked_image[mask!=0]=[0,0,0]
cv2_imshow(masked_image)
```

```python
scene=cv2.imread('/content/drive/MyDrive/Computer vision Images/space.jpg')
scene=cv2.cvtColor(scene,cv2.COLOR_BGR2RGB)
```

```python
dim=(540,360)
img_resized=cv2.resize(scene,dim)
cv2_imshow(img_resized)
```

```python
img_resized[mask==0]=[0,0,0]
cv2_imshow(img_resized)
```

```python
man_in_space=img_resized+masked_image
cv2_imshow(man_in_space)
```
